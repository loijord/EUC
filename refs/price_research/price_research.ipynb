# What's the best range for cheapest price?

No doubt, the winner is one of those with **Gotway** brand. But which one?
* Take into account you need a huge wheel in order to be able to climb downstairs/upstairs and be safe on bumpy roads
* Take into account you don't want a weight a lot more than 30 kilos because it just wastes a battery and make it impossible to bring
* Personally, I don't think you really need a speed higher than 50km/h or high power (high torque) because we are not living in the republic of mountains

So...

* If you are on a budget (Big no to more than 1k) then Gotway MCM5 is a clever choice (price: 1050€, range: 70km, wheel: 14 inches)
* If you say no to ranges lower than 60km but still on a budget, buy Begode Tesla V2 (price: 1290€, range: 90km, wheel: 16 inches) or Begode Tesla V3 (price: 1600€, range: 130km, wheel: 16inches)
* If you want extremely huge range and still feel safe on a road (it means, huge wheel), not to spend 1k just for the ,,good brand name" (haha), buy Gotway Nikola Plus (price: 1990€, range: 170km, wheel: 17 inches) or Gotway EXN 19`` (price: 2600€, range: 200km, wheel: 20 inches)

### Range: km/eur vs wheel size (so: safety)

```python
import plotly.graph_objects as go

brand_coloring = {'Kingsong':'red', 'Inmotion':'green', 'Gotway':'purple', 'Leaperkim':'orange', 'Ninebot':'yellow'}
wheel_coloring = {10, 11, 14, 16, 17, 18, 19, 20, 22, 24}

fig1 = go.Scatter(x=df['range'], y=df['price']/df['range'], mode='markers', 
                  text=[f'model: {model}<br>price: {price} <br>wheel: {wheel}' for model, price, wheel in df[['model', 'price', 'wheel']].values],
                  marker_color=[brand_coloring[n] for n in df['brand']],
                  marker_size= 1.5* df['wheel'])
    
fig = go.Figure(data=[fig1])
fig.update_layout(xaxis_title="Range", 
                  yaxis_title='Price', 
                  width=1000)

path = f'price_range_research_plots/vs_wheel.html'
with open(path, "w", encoding="utf-8") as f: 
    f.write(fig.to_html())
```
